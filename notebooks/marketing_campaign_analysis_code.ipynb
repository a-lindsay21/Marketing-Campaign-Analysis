```python
import pandas as pd

# Load the uploaded dataset
df = pd.read_csv('marketing_campaign_dataset.csv')

# Display the first few rows
print(df.head())

# Get dataset information
print(df.info())

# Check for missing values
print("Missing values per column:\n", df.isnull().sum())

# Check for duplicate rows
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Example: Drop rows with missing values
df = df.dropna()
```

```python
# Calculate engagement rate
df['Engagement_Rate'] = (df['Clicks'] / df['Impressions']) * 100
print(f"Average Engagement Rate: {df['Engagement_Rate'].mean():.2f}%")
```

```python
# Response rate by channel
channel_effectiveness = df.groupby('Channel_Used')['Conversion_Rate'].mean() * 100
print(channel_effectiveness)
```

```python
roi_by_channel = df.groupby('Channel_Used')['ROI'].mean()
print(roi_by_channel)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart for channel effectiveness
plt.figure(figsize=(8, 6))
sns.barplot(x=channel_effectiveness.index, y=channel_effectiveness.values, palette='pastel')
plt.title('Conversion Rate by Channel')
plt.ylabel('Conversion Rate (%)')
plt.xlabel('Channel')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=roi_by_channel.index, y=roi_by_channel.values, palette='muted')
plt.title('ROI by Channel')
plt.ylabel('ROI')
plt.xlabel('Channel')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Customer_Segment', y='Engagement_Score', data=df, palette='coolwarm')
plt.title('Engagement Score by Customer Segment')
plt.xticks(rotation=45)
plt.show()
```
